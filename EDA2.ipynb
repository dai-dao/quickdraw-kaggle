```python
import pandas as pd
import os
import glob
import time
```

```python
# Path definitions
base_dir = 'input'
test_path = os.path.join('input', 'test_simplified.csv')
ALL_TRAIN_PATHS = glob.glob(os.path.join(base_dir, 'train_processed/*.csv'))
```

```python
all_dfs = []

def read_df(p):
    # print(os.path.basename(p))
    all_dfs.append(pd.read_csv(p))
    print(len(all_dfs))

for p in ALL_TRAIN_PATHS[170:]:
    read_df(p)
```

```python
big_df = pd.concat(all_dfs)
print(big_df.shape)
```
